# 08 — Evaluation

Compute all metrics and inspect results before extracting to `ave/evaluation/`.

**Covers:**
- Load predictions from all variants (`data/WDC-PAVE/predictions/`)
- Valid JSON rate and schema adherence
- Field-level precision / recall / F1 (case-insensitive, set-level for multi-value)
- Multi-value F1 (separate metric for the 3.9% multi-value attributes)
- Object exact match
- Hallucination rate, miss rate, null accuracy
- Latency and cost-per-1k stats
- Error bucketing: wrong value, partial match, hallucination, formatting, normalization mismatch
